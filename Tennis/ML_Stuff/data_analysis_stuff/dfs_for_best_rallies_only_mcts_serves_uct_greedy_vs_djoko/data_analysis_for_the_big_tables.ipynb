```python
import pandas as pd
import ast
```

```python
df = pd.read_csv(r"C:\Users\Carlo\Repos\TrainingsTool\Tennis\ML_Stuff\data_analysis_stuff\dfs_for_best_rallies_only_mcts_serves_uct_greedy_vs_djoko\df_2nd_serves_ad_side.csv", encoding = 'unicode_escape', low_memory=False)

df['Ralley'] = df['Ralley'].apply(ast.literal_eval)
```

```python
df["serve_direction"] = "-"
df["return_direction"] = "-"
df["third_shot_direction"] = "-"
```

```python
# split the dataframe into 3 seperate frames, one for each first serve direction
for index, row in df.iterrows():
    
    if any("4" in value for value in row['Ralley'][1]):
        df["serve_direction"][index] = "4"
    elif any("5" in value for value in row['Ralley'][1]):
        df["serve_direction"][index] = "5"
    elif any("6" in value for value in row['Ralley'][1]):
        df["serve_direction"][index] = "6"
    
    if any("1" in value for value in row['Ralley'][2]):
        df["return_direction"][index] = "1"
    elif any("2" in value for value in row['Ralley'][2]):
        df["return_direction"][index] = "2"
    elif any("3" in value for value in row['Ralley'][2]):
        df["return_direction"][index] = "3"

    
    if any("1" in value for value in row['Ralley'][3]):
        df["third_shot_direction"][index] = "1"
    elif any("2" in value for value in row['Ralley'][3]):
        df["third_shot_direction"][index] = "2"
    elif any("3" in value for value in row['Ralley'][3]):
        df["third_shot_direction"][index] = "3"
```

```python
df_serve_4 = df[df['serve_direction'] == '4']
df_serve_5 = df[df['serve_direction'] == '5']
df_serve_6 = df[df['serve_direction'] == '6']
```
