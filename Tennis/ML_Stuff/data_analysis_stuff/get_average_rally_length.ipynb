```python
import pandas as pd
import ast
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\Carlo\Repos\TrainingsTool\Tennis\exported_results\mcts_vs_average\mcts_uct_uct_vs_average_101_99.csv", encoding='unicode_escape', low_memory=False)
```

```python
len(df)
```

```python
df_test = df
df_test
```

```python
# Convert string representations of lists to actual lists
df_test['Ralley'] = df_test['Ralley'].apply(ast.literal_eval)

# Calculate the length of each array and add a new column 'Ralley_Length'
df_test['Ralley_Length'] = df_test['Ralley'].apply(lambda x: len(x) if isinstance(x, list) else 0)

# Calculate the average length
average_length = df_test['Ralley_Length'].mean()

# Print the DataFrame with the new column
print(df_test)

# Print the average length
print("Average Ralley Length:", average_length)
```

```python
df_test
```

```python
# Calculate the standard deviation of 'Ralley_Length'
std_deviation = df_test['Ralley_Length'].std()

# Print the standard deviation
print("Standard Deviation of Ralley_Length:", std_deviation)
```

```python

# Calculate the mean and standard deviation
average_length = df_test['Ralley_Length'].mean()
std_deviation = df_test['Ralley_Length'].std()

# Plotting
plt.bar(['Average Length'], [average_length], yerr=[std_deviation], capsize=5, color='blue')
plt.title('Average Length of Ralley')
plt.ylabel('Length')
plt.show()
```
