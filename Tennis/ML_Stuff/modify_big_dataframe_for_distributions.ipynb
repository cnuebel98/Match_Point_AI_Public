```python
def process_row(row):
    # Filter out NaN values and convert to a list of strings
    values = [str(value) for value in row.dropna()]
    
    # Add double quotes to strings
    values = ['"' + val + '"' if isinstance(val, str) else val for val in values]
    
    # Convert the list to a string representation
    result = '[' + ','.join(values) + ']'
    
    return result

# Apply the function to each row
new_df = df_djoko.apply(process_row, axis=1)

# Create a new DataFrame with the processed values
new_df = pd.DataFrame(new_df, columns=['processed_values'])

# Display the result
print(new_df)
```
