```python
import pandas as pd
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(df['Point_Winner'], df['C_Value'])

# Print the contingency table
print("Contingency Table:")
print(contingency_table)

# Perform chi-squared test
chi2, p, _, _ = chi2_contingency(contingency_table)

# Print the results
print("\nChi-squared value:", chi2)
print("P-value:", p)

# Check for significance
alpha = 0.05
if p < alpha:
    print("\nThe result is significant. There is a significant relationship between Point_Winner and C_Value.")
else:
    print("\nThe result is not significant. There is no significant relationship between Point_Winner and C_Value.")
```

```python
import pandas as pd
import numpy as np
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from sklearn.preprocessing import LabelEncoder

# Encode categorical variables to numerical values
label_encoder = LabelEncoder()
df['Point_Winner'] = label_encoder.fit_transform(df['Point_Winner'])
df['C_Value'] = label_encoder.fit_transform(df['C_Value'])

# Perform pairwise comparisons with Bonferroni-Holm correction
tukey_result = pairwise_tukeyhsd(endog=df['Point_Winner'], groups=df['C_Value'], alpha=0.05)

# Extract groups 1 and 2
groups_01 = df[df['C_Value'].isin([1, 2])]

# Create a contingency table
contingency_table = pd.crosstab(groups_01['Point_Winner'], groups_01['C_Value'])

# Calculate Cramér's V
n = contingency_table.sum().sum()
r, c = contingency_table.shape
chi2 = contingency_table.apply(lambda x: (x - x.mean()) ** 2 / x.mean()).sum().sum()
cramers_v = np.sqrt(chi2 / (n * min((r-1), (c-1))))

print("Cramer's V for groups 0 and 1:", cramers_v)
```
